```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
validating = requests.get('https://www.imdb.com/chart/toptv/?sort=ir,desc&mode=simple&page=1')
validating.status_code
```

```python
# Parsing Content from url
tv_shows = BeautifulSoup(validating.text,'html.parser')
```

```python
Show_name= tv_shows.find_all('td',{'class':'titleColumn'})
rating= tv_shows.find_all('td',{'class':'ratingColumn imdbRating'})
name =[]
year=[]
show_rating=[]
user_rated_that_movie=[]
url=[]
top_actors=[]
for i in range(len(Show_name)):
    name.append(Show_name[i].text.strip().replace(" ","").split('\n')[1])
    year.append(Show_name[i].text.strip().replace(" ","").split('\n')[2].replace('(',"").replace(')',""))
    show_rating.append(rating[i].text.strip())
    for body in tv_shows.find_all('td',{'class':'ratingColumn imdbRating'}):
            if body.find('strong'):
                text= body.find('strong')
                title= text.get('title')
                x= title.split(" ")
                user_rated_that_movie.append(x[3])
    for body in tv_shows.find_all('td',{'class':'titleColumn'}):
            if body.find('a'):
                text= body.find('a')
                top_actor= text.get('title')
                #text_link=body.find('a')['href']
                #text_link='https://www.imdb.com/'+ text_link +'?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=12230b0e-0e00-43ed-9e59-8d5353703cce&pf_rd_r=BE897XJHS7BYFMFFPCRK&pf_rd_s=center-1&pf_rd_t=15506&pf_rd_i=toptv&ref_=chttvtp_tt_2'
                top_actors.append(top_actor)
```

```python
shows_data=pd.DataFrame(list(zip(name,year,show_rating,user_rated_that_movie,top_actors)),columns=['Name','Release year','Show_rating','No_users_rated','Top actors'])
shows_data
```
